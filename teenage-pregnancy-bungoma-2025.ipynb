```python
# 📘 Teenage Pregnancy in Bungoma County, Kenya (2025)

# 🧾 Introduction

"""
Teenage pregnancy is a critical public health and social issue affecting thousands of young girls across Kenya. Bungoma County, located in the Western region of Kenya, has seen rising rates of teenage pregnancies, which are linked to school dropouts, early marriages, and long-term economic challenges.

This project aims to analyze the teenage pregnancy situation in Bungoma County for the year 2025. Using simulated data that includes all sub-counties, wards, and villages, we'll identify trends, patterns, and relationships between various factors such as access to healthcare, education programs, and contraceptive availability.
"""

# 📦 Data Loading and Exploration

import pandas as pd

# Load the dataset
df = pd.read_csv("/kaggle/input/teenage-pregnancy-in-bungoma-county-2025/bungoma.csv")

# Preview the data
print("First five rows of the dataset:")
display(df.head())

# Check data info
print("\nDataset Info:")
df.info()

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Unique sub-counties
print("\nUnique Sub-counties:")
print(df['SubCounty'].unique())

# Summary statistics
print("\nSummary Statistics:")
print(df.describe())

# 📊 Data Visualization

import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

# Set plot style
sns.set(style="whitegrid")

# Teenage Pregnancy Rate by Sub-County
plt.figure(figsize=(14,6))
sns.barplot(x='SubCounty', y='TeenPregnancyRate', data=df, ci=None, estimator='mean', palette="coolwarm")
plt.title('Average Teenage Pregnancy Rate by SubCounty (2025)', fontsize=14)
plt.xticks(rotation=45)
plt.ylabel("Teen Pregnancy Rate (%)")
plt.tight_layout()
plt.show()

# School Dropouts Distribution
plt.figure(figsize=(10,6))
sns.histplot(df['SchoolDropouts'], kde=True, bins=30, color='coral')
plt.title("Distribution of School Dropouts", fontsize=14)
plt.xlabel("Number of Dropouts")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# Pregnancy rate vs. contraceptive access
plt.figure(figsize=(8,6))
sns.boxplot(x='ContraceptiveAccess', y='TeenPregnancyRate', data=df)
plt.title("Teen Pregnancy Rate vs Contraceptive Access", fontsize=14)
plt.tight_layout()
plt.show()

# Pregnancy rate vs. education programs
plt.figure(figsize=(8,6))
sns.boxplot(x='EducationPrograms', y='TeenPregnancyRate', data=df)
plt.title("Teen Pregnancy Rate vs Education Programs", fontsize=14)
plt.tight_layout()
plt.show()

# 🔍 Correlation Analysis

# Encode binary fields
df_encoded = df.copy()
df_encoded['ContraceptiveAccess'] = df_encoded['ContraceptiveAccess'].map({'Yes': 1, 'No': 0})
df_encoded['EducationPrograms'] = df_encoded['EducationPrograms'].map({'Yes': 1, 'No': 0})

# Correlation matrix
plt.figure(figsize=(10,8))
from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy()
for column in df_encoded.columns:
    if df_encoded[column].dtype == 'object':
        df_encoded[column] = LabelEncoder().fit_transform(df_encoded[column])

sns.heatmap(df_encoded.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numeric Variables', fontsize=14)
plt.tight_layout()
plt.show()

# 📝 Observations

"""
- Sub-counties such as Kanduyi and Bumula have higher average teenage pregnancy rates.
- Villages with lower access to contraceptives show significantly higher pregnancy rates.
- Education programs appear to be effective in reducing pregnancy rates when available.
- Areas with more health centers tend to have fewer dropouts.
- A strong correlation is noted between school dropouts and teenage pregnancy rates.
"""

# ✅ Recommendations

"""
1. **Expand Sexual Education Programs**: Roll out curriculum-based and community-led reproductive education across all wards and villages.
2. **Improve Contraceptive Access**: Ensure contraceptives are available in all health centers and through community outreach.
3. **Strengthen Health Infrastructure**: Invest in more health facilities and maternal support services in underserved areas.
4. **Track Dropout Causes**: Implement a county-level education tracking system to respond quickly to dropouts caused by pregnancy.
5. **Engage Community Leaders**: Partner with elders and religious leaders to promote behavior change and delay early marriages.
"""

# 📌 Conclusion

"""
This analysis provides a data-driven understanding of teenage pregnancy trends across Bungoma County. Although the data is simulated for learning purposes, the insights can guide real-world policy discussions. By improving access to contraceptives, enhancing health services, and implementing effective educational programs, Bungoma County can reduce the incidence of teenage pregnancies and their associated socio-economic impacts.
"""

```
